```python
from sklearn.datasets import fetch_california_housing
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
california_housing = fetch_california_housing(as_frame=True)
print(california_housing.DESCR)
df = california_housing.frame
print(df.head())
df.info()
```

```python
df["MedHouseVal"] = pd.qcut(df["MedHouseVal"], 6, retbins=False)
df["MedHouseVal"] = df["MedHouseVal"].apply(lambda x: x.mid)
sns.pairplot(data=df, hue="MedHouseVal")
```

```python
corr = df.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

```python
df.hist(alpha=0.5, figsize=(20, 10))
```

Zmienne AveBedrms, AveRooms, Population i AveOccup wszystkie mają bardzo skupione wartości. Prawie wszystkie domostwa maja te same wartości dla zadanych zmiennych.

```python
df = df.loc[(df["AveBedrms"] < 4) & (df["AveRooms"] < 21) & (df["AveOccup"] <
9) & (df["Population"] < 8000)]
```

```python
df.hist(alpha=0.5, figsize=(20, 10))
```

Zmienne które wcześniej były skupione są dokłądniejsze na wykresach co pozwala odczytać konkretniejsze wartości.
Podczas regresji najważniejsze będzie MedInc

```python
MedHouseVal = df.loc[:, "MedHouseVal"]
df = df.drop("MedHouseVal", axis=1)

df.info()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df, MedHouseVal, test_size=0.2, random_state = 0)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

for column in X_train.columns:
    X_train[column] = scaler.fit_transform(X_train[column].to_numpy().reshape(-1, 1))
    X_test[column] = scaler.transform(X_test[column].to_numpy().reshape(-1, 1))
    
y_train = pd.DataFrame(scaler.fit_transform(y_train.to_numpy().reshape(-1, 1)), columns = ["MedHouseVal"])
y_test = pd.DataFrame(scaler.transform(y_test.to_numpy().reshape(-1, 1)), columns = ["MedHouseVal"])
```

```python
X_train.hist(alpha=0.5, figsize=(20, 10))
y_train.hist(alpha=0.5, figsize=(20, 10))
#tak jak wcześniej tylko przeskalowane
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

model = LinearRegression().fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print(mean_absolute_error(y_test, y_pred))

print(y_pred.mean(), y_test.mean())
print("Score: ", model.score(X_test, y_test))
# Średnie wartości są bardzo zbliżone
```

Średnie wartości są bardzo zbliżone, w stosunku do średnich wartości jest dość spory.

```python
x_pred_train = model.predict(X_train)
print(mean_absolute_error(x_pred_train, y_train))
# Model nie dopasował się idealnie do danych treningowych więc napewno nie będzie overfittingu, występuje dość znaczny błąd co może oznaczać underfitting
```

```python
from matplotlib import pyplot
coef = model.coef_
print(coef[0])
pyplot.bar([x for x in range(len(coef[0]))], coef[0])
pyplot.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=3, include_bias=False)

X_train2 = poly_features.fit_transform(X_train)
X_test2 = poly_features.transform(X_test)
```

```python
print(X_test2.shape)
```

```python
model2 = LinearRegression().fit(X_train2, y_train)
```

```python
y_pred = model2.predict(X_test2)
print(mean_absolute_error(y_test, y_pred))

print(y_pred.mean(), y_test.mean())
print("Score: ", model2.score(X_test2, y_test))
```

```python
poly_features = PolynomialFeatures(degree=5, include_bias=False)

X_train3 = poly_features.fit_transform(X_train)
X_test3 = poly_features.transform(X_test)

print(X_test3.shape)
model3 = LinearRegression().fit(X_train3, y_train)

y_pred = model3.predict(X_test3)
print(mean_absolute_error(y_test, y_pred))

print(y_pred.mean(), y_test.mean())
print("Score: ", model3.score(X_test3, y_test))
```

Wartości poprawiły się dla degree = 3. W sytuacji degree = 5 znacznie się pogarszają.

```python
coef = model3.coef_
pyplot.bar([x for x in range(len(coef[0]))], coef[0])
pyplot.show()
```

```python
import numpy as np
print(max(coef[0]))
print(min(coef[0]))
print(coef[0].size - np.count_nonzero(coef[0]))
```

Brak wag zero

```python
from sklearn.linear_model import Ridge, Lasso

rdg = Ridge(random_state=0)
las = Lasso(random_state=0)

rdg.fit(X_train2, y_train)
print(mean_absolute_error(y_test, rdg.predict(X_test2)))

coef = rdg.coef_
pyplot.bar([x for x in range(len(coef[0]))], coef[0])
pyplot.show()
print(max(coef[0]))
print(min(coef[0]))
print(coef[0].size - np.count_nonzero(coef[0]))

rdg.fit(X_train3, y_train)
print(mean_absolute_error(y_test, rdg.predict(X_test3)))

coef = rdg.coef_
pyplot.bar([x for x in range(len(coef[0]))], coef[0])
pyplot.show()
print(max(coef[0]))
print(min(coef[0]))
print(coef[0].size - np.count_nonzero(coef[0]))

las.fit(X_train2, y_train)
print(mean_absolute_error(y_test, las.predict(X_test2)))

las.fit(X_train3, y_train)
print(mean_absolute_error(y_test, las.predict(X_test3)))
```

Nie pojawiają sie zerowe wagi, model Ridge radzi sobie znacznie lepiej.

```python
from sklearn.linear_model import LassoCV

# parameter cv for number of folds in cross validation

lasCV = LassoCV(cv = 5, random_state=0)

lasCV.fit(X_train2, y_train)
```

```python
coef = lasCV.coef_
pyplot.bar([x for x in range(len(coef))], coef)
pyplot.show()
print(max(coef))
print(min(coef))
print(coef[0].size - np.count_nonzero(coef))
print(mean_absolute_error(y_test, lasCV.predict(X_test2)))

print(f'Wyeliminowano {round(100 * 50 / len(coef))}% cech')
```

Najlepsze wyniki otrzymały: model LassoCV przy zbiorze rozszerzonym (degree=3), Ridge przy zbiorze rozszerzonym (degree=5)

```python
import pandas as pd

train_data = pd.read_csv("penguins.csv")
train_data.head()
```

```python
train_data= train_data.dropna()
train_data = train_data.drop(["year"], axis=1)
train_data.info()
```

```python
print(train_data["species"].value_counts())
```

Dane są niezbalansowane

```python
import seaborn as sns
sns.pairplot(data=train_data, hue="species")
```

```python
corr = train_data.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

Pingwiny Gento są pod większością względów większe od pozostałych, a Adelie mniejsze.
Silna korelacja pomiędzy długością dzioba i długością płetw.

```python
ad_ch = train_data.loc[(train_data["species"] == "Adelie") | (train_data["species"] == "Chinstrap")]
islands = ad_ch["island"].unique()
print(islands)
```

```python
ch = ad_ch.loc[(ad_ch["species"] == "Chinstrap")]
print(ch["island"].unique())
```

```python
ad_ch = ad_ch.drop(columns=["island"])
```

Głównie rozróżnia nasze klasy cecha długości dzioba.

```python
ad_ch["sex"] = ad_ch["sex"].map({"male": 1, "female": 0})
```

```python
ad_ch.head()
```

```python
species = ad_ch.pop("species")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(ad_ch, species, test_size=0.2, stratify=species, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
cols = ["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]

X_train[cols] = scaler.fit_transform(X_train[cols])
X_test[cols] = scaler.transform(X_test[cols])
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(class_weight="balanced", solver="lbfgs", random_state=0)

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
res = (y_test != y_pred).sum()

print("Liczba źle sklasyfikowanych:", res, ", procent:", round(100*res/y_test.size), "%")
```

W zależności od tego jak podzialiły się zbiory testowy i treningowy wyniki wachają się od dobrych 1 błąd do bardzo złych 14 błędów. (Raczej dostajemy dobre wyniki ale przy pierwszej próbie z nieznanego powodu było 14 błędów - wszystko zaklasyfikował jako Adelie)

```python
weights = pd.DataFrame({"feature": X_train.columns, "coef": model.coef_[0]})
print(weights)
```

Zdecydowanie najwazniejsza jest długość dzioba co zgadza się z przewidywaniami.

```python
X_train1 = X_train.drop(columns=["bill_length_mm"])
X_test1 = X_test.drop(columns=["bill_length_mm"])
```

```python
model1 = LogisticRegression(class_weight="balanced", solver="lbfgs", random_state=0)

model1.fit(X_train1, y_train)

y_pred = model1.predict(X_test1)
res = (y_test != y_pred).sum()

print("Liczba źle sklasyfikowanych:", res, ", procent:", round(100*res/y_test.size), "%")
```

Dużo większy błąd, bo najważniejsza cecha zniknęła.

```python
X_new = X_train[["bill_depth_mm", "flipper_length_mm"]]
X_test_new = X_test[["bill_depth_mm", "flipper_length_mm"]]

model2 = LogisticRegression(class_weight="balanced", solver="lbfgs", random_state=0)

model2.fit(X_new, y_train)

y_pred = model2.predict(X_test_new)
res = (y_test != y_pred).sum()

print("Liczba źle sklasyfikowanych:", res, ", procent:", round(100*res/y_test.size), "%")
```

```python
weights = pd.DataFrame({"feature": X_new.columns, "coef": model2.coef_[0]})
print(weights)
```

Wybrełem głębokośc dzioba i długośćpłetwy przez macierz korelacji. Jak się można było spodziawać wyniki są gorsza ale 60% poprawności nie jast złym wynikiem.

```python
scores = np.dot(X_test_new, model2.coef_[0]) + model2.intercept_

predictions = ["Chinstrap" if score >= 0 else "Adelie" for score in scores]
predictions2 = model2.predict(X_test_new)

print(predictions, predictions2)
```
