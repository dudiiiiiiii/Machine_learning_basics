```python
import gzip
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def load_data():
    base_url = "http://fashion-mnist.s3-website.eu-central-1.amazonaws.com/"
    files = ["train-images-idx3-ubyte.gz", "train-labels-idx1-ubyte.gz"]
    
    with gzip.open(files[0], "rb") as f:
        train_images = np.frombuffer(f.read(), np.uint8, offset=16).reshape(-1, 28, 28)
        
    with gzip.open(files[1], "rb") as f:
        train_labels = np.frombuffer(f.read(), np.uint8, offset=8)
        
    return train_images, train_labels
```

```python
train_images, train_labels = load_data()
```

```python
def extract(images, labels, cls_lab, num):
    photos = []
    for i in range(len(images)):
        if labels[i] == cls_lab:
            photos.append(images[i])
        if len(photos) == num:
            break
    return photos
```

```python
tshirt_top = extract(train_images, train_labels, 0, 20)
pullover = extract(train_images, train_labels, 2, 20)
dress = extract(train_images, train_labels, 3, 20)
data = tshirt_top+pullover+dress
```

```python
plt.figure(figsize=(10, 10))

for i in range(20):
    plt.subplot(5, 12, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(tshirt_top[i], cmap=plt.cm.binary)
    plt.xlabel("Tshirt_top")
    
    plt.subplot(5, 12, i+21)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(pullover[i], cmap=plt.cm.binary)
    plt.xlabel("Pullover")
    
    plt.subplot(5, 12, i+41)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(dress[i], cmap=plt.cm.binary)
    plt.xlabel("Coat")

plt.show()
```

```python
tmp = np.array(data)
tmp = tmp.reshape(60, 784)

tmp = tmp.astype(np.float) / 255.0
mean = np.mean(tmp, axis=0)
centered = tmp - mean
```

```python
cov = np.cov(centered, rowvar=False)
var = np.var(centered, axis=0)

print(cov.shape)
print(var.shape)
```

```python
from sklearn.decomposition import PCA

new = centered
pca = PCA()
pca.fit(new)
```

```python
principal_components = pca.components_  # Nowe wektory bazowe
explained_variance = pca.explained_variance_  # Powiązane wariancje
print(principal_components.shape, explained_variance.shape)
```

1. Macierz cowariancji przed to symetrycznamacierz 784x784 każda cecha to jeden piksel.

```python
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.bar(range(1, 785), var, align='center')
plt.xlabel('Numer cechy')
plt.ylabel('Wariancja')
plt.title('Wariancje przed transformacją PCA')
plt.subplot(1, 2, 2)
plt.bar(range(1, 61), explained_variance, align='center')
plt.xlabel('Numer cechy')
plt.ylabel('Wariancja')
plt.title('Wariancje po transformacji PCA')
plt.show()
```

```python
mean_img = mean.reshape(28,28)

plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.imshow(mean_img, cmap='gray')
plt.title('Średnie zdjęcie')
plt.show()
```

Widac że średnie zdjęcie ma zarys elementu garderoby.


```python
# Posortowanie wektorów bazowych według powiązanej wariancji
sorted_idx = np.argsort(explained_variance)[::-1]  # Indeksy posortowane malejąco
principal_components_sorted = principal_components[sorted_idx]

print(principal_components_sorted.shape)
```

```python
plt.imshow(principal_components_sorted, cmap='hot', aspect='auto')
plt.colorbar()  # Add colorbar for reference
plt.title('Heatmap of sorted matrix vectors')
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.show()
```

```python
n_components = principal_components.shape[0]
for i in range(60):
    plt.figure(figsize=(3, 3))
    plt.imshow(principal_components_sorted[i].reshape(28, 28), cmap='gray')  # Obraz z wektora bazowego
    plt.show()
```

Zancznie mniejsza dokładność obrazków.

```python
n_components_list = [3, 9, 27]
X_centered = centered
X_mean = mean

for n_components in n_components_list:
    # Wyzerowanie wartości cech poza wybraną garstką
    X_reduced = np.dot(X_centered, pca.components_[:n_components].T)  # Konwersja do nowej bazy
    X_reduced[:, n_components:] = 0  # Wyzerowanie zbędnych cech

    # Przywrócenie do oryginalnej bazy
    X_restored = np.dot(X_reduced, pca.components_[:n_components])  # Konwersja do oryginalnej bazy
    X_restored += X_mean  # Dodanie średniego wektora

    # Przekształcenie wektorów cech na obrazy
    X_restored = np.clip(X_restored, 0, 1)  # Ograniczenie wartości pikseli do przedziału [0, 1]
    X_restored = (X_restored * 255).astype(np.uint8)  # Przeskalowanie wartości pikseli do zakresu [0, 255]
    X_restored = X_restored.reshape(-1, 28, 28)  # Przekształcenie wektorów cech na obrazy

    # Wyświetlenie "odchudzonych" fotografii
    plt.figure(figsize=(3,3))
    plt.imshow(X_restored[0], cmap='gray')  # Wyświetlenie obrazu
    plt.title(f'PC {n_components}')  # Numer głównej składowej
    plt.show()  
        
    
```

Wraz z rosnącą liczbą komponentów rośnie wyraźność zdjęcia.

```python
pca = PCA(n_components=2)
X_2d = pca.fit_transform(X_centered)

# Utworzenie słownika z kolorami odpowiadającymi różnym typom obiektów
color_dict = {'sukienka': 'blue', 'polar': 'red', 'tshirt': 'green'}

# Utworzenie listy kolorów na podstawie typów obiektów
colors = ['green' for i in range(20)] + ['red' for i in range(20)] + ['blue' for _ in range(20)]

# Wygenerowanie wykresu z kropkami na płaszczyźnie 2D
plt.scatter(X_2d[:, 0], X_2d[:, 1], c=colors, alpha=0.5)
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('Wizualizacja zbioru na płaszczyźnie 2D z uwzględnieniem typów obiektów')
plt.show()
```
