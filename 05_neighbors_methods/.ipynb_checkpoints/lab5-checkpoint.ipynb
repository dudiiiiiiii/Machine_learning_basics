## Część I: obrazy jako cechy

Obrazy z dwóch wybranych klas
- 0 - samoloty
- 1 - samochody

```python
# Załaduj zbiór danych CIFAR-10
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()

# Wybierz dwie wybrane klasy (np. samochody i psy)
selected_classes = [0, 1]  # Indeksy klas - samochody: 1, psy: 5

# Wybierz obrazy i etykiety z wybranych klas
train_images = x_train[np.where(np.isin(y_train, selected_classes))[0]]
train_labels = y_train[np.where(np.isin(y_train, selected_classes))[0]]
test_images = x_test[np.where(np.isin(y_test, selected_classes))[0]]
test_labels = y_test[np.where(np.isin(y_test, selected_classes))[0]]

# Wyodrębnij obrazy i etykiety dla wybranych klas
class1_images = train_images[train_labels.flatten() == selected_classes[0]]
class1_labels = train_labels[train_labels.flatten() == selected_classes[0]]

class2_images = train_images[train_labels.flatten() == selected_classes[1]]
class2_labels = train_labels[train_labels.flatten() == selected_classes[1]]

# Wyświetl kilka obrazów z obu klas
plt.figure(figsize=(10, 4))
for i in range(5):
    plt.subplot(2, 5, i+1)
    plt.imshow(class1_images[i])
    plt.title(f'Class: {class1_labels[i]}')
    plt.axis('off')

    plt.subplot(2, 5, i+6)
    plt.imshow(class2_images[i])
    plt.title(f'Class: {class2_labels[i]}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from imblearn.over_sampling import SMOTE
from sklearn.metrics import accuracy_score, classification_report
import cv2
import imutils

def transform_images_to_vector(images):
    vectors = []
    for image in images:
        vectors.append(image.reshape(-1))
    return np.array(vectors)

def extract_color_histogram(images, bins=(8, 8, 8)):
    vectors = []

    for image in images:
        # extract a 3D color histogram from the HSV color space using
        # the supplied number of `bins` per channel
        hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
        hist = cv2.calcHist([hsv], [0, 1, 2], None, bins,
          [0, 180, 0, 256, 0, 256])
        # handle normalizing the histogram if we are using OpenCV 2.4.X
        if imutils.is_cv2():
          hist = cv2.normalize(hist)
        # otherwise, perform "in place" normalization in OpenCV 3 (I
        # personally hate the way this is done
        else:
          cv2.normalize(hist, hist)
        # return the flattened histogram as the feature vector
        vectors.append(hist.flatten())
    return np.array(vectors)
```

```python
# 1. Przekształcenie obrazów na wektory
train_images_vector = transform_images_to_vector(train_images)
test_images_vector = transform_images_to_vector(test_images)

# 2. Przekształcenie obrazów na histogramy
train_images_histograms = extract_color_histogram(train_images)
test_images_histograms = extract_color_histogram(test_images)
```

```python
# Algorytm kNN na wektorach obrazów
def kNN_classification_vector(train_images, train_labels, test_images, test_labels, k):
    knn = KNeighborsClassifier(n_neighbors=k, n_jobs=-1)
    knn.fit(train_images, train_labels)
    accuracy = knn.score(test_images, test_labels)
    return accuracy

# Algorytm kNN na histogramach obrazów
def kNN_classification_histogram(train_images, train_labels, test_images, test_labels, k):
    knn = KNeighborsClassifier(n_neighbors=k, n_jobs=-1)
    knn.fit(train_images, train_labels)
    accuracy = knn.score(test_images, test_labels)
    return accuracy
```

```python
# Klasyfikacja kNN dla różnych liczby sąsiadów i różnych podejść
k_values = [1, 3, 5, 7, 9]
approach_vector_accuracy = []
approach_histogram_accuracy = []

for k in k_values:
    print(k)
    vector_accuracy = kNN_classification_vector(train_images_vector, train_labels, test_images_vector, test_labels, k)
    histogram_accuracy = kNN_classification_histogram(train_images_histograms, train_labels, test_images_histograms, test_labels, k)
    approach_vector_accuracy.append(vector_accuracy)
    approach_histogram_accuracy.append(histogram_accuracy)
```

```python
# Najlepsze wyniki
best_vector_accuracy = max(approach_vector_accuracy)
best_vector_k = k_values[approach_vector_accuracy.index(best_vector_accuracy)]

best_histogram_accuracy = max(approach_histogram_accuracy)
best_histogram_k = k_values[approach_histogram_accuracy.index(best_histogram_accuracy)]

print("Najlepsze wyniki dla podejścia z wektorami:")
print("Liczba sąsiadów:", best_vector_k)
print("Dokładność klasyfikacji:", best_vector_accuracy)

print("\nNajlepsze wyniki dla podejścia z histogramami:")
print("Liczba sąsiadów:", best_histogram_k)
print("Dokładność klasyfikacji:", best_histogram_accuracy)

# Klasyfikacja kNN dla najlepszego podejścia
best_approach = 'vector' if best_vector_accuracy > best_histogram_accuracy else 'histogram'
best_k = best_vector_k if best_approach == 'vector' else best_histogram_k

if best_approach == 'vector':
    train_features = train_images_vector
    test_features = test_images_vector
else:
    train_features = train_images_histograms
    test_features = test_images_histograms
```

```python
knn = KNeighborsClassifier(n_neighbors=best_k, n_jobs=-1)
knn.fit(train_features, train_labels)
_, indices = knn.kneighbors(test_features)
predicted_labels = train_labels[indices]

# Wyświetlenie sąsiadów dla przykładów z obu klas
class1_indices = np.where(test_labels.flatten() == 0)[0][:3]
class2_indices = np.where(test_labels.flatten() == 1)[0][:3]

class1_neighbors = train_images[np.squeeze(indices[class1_indices])]
class2_neighbors = train_images[np.squeeze(indices[class2_indices])]

# Wypisanie wyników
print("\nPrzykłady z klasy 1 i ich sąsiedzi:")
for i in range(3):
    print("Przykład", i+1)
    plt.figure(figsize=(14, 4))
    plt.subplot(1, best_k+1, 1)
    plt.imshow(test_images[class1_indices[i]])
    plt.title("Przykład testowy")
    plt.axis('off')
    for j in range(best_k):
        plt.subplot(1, best_k+1, j+2)
        plt.imshow(class1_neighbors[i][j])
        plt.title(f"Sąsiad {j+1}")
        plt.axis('off')
    plt.show()

print("\nPrzykłady z klasy 2 i ich sąsiedzi:")
for i in range(3):
    print("Przykład", i+1)
    plt.figure(figsize=(14, 4))
    plt.subplot(1, best_k+1, 1)
    plt.imshow(test_images[class2_indices[i]])
    plt.title("Przykład testowy")
    plt.axis('off')
    for j in range(best_k):
        plt.subplot(1, best_k+1, j+2)
        plt.imshow(class2_neighbors[i][j])
        plt.title(f"Sąsiad {j+1}")
        plt.axis('off')
    plt.show()
```

Wyniki mimo, że nie zawsze są poprawne to zawsze wydają się być bardzo sensowne. Widać że algorymy skupiły się przede wszystkim na aspektach kolorystycznych (oprócz klasa 1 przykład 3 gdzie mogło być brane bardziej ułożenie obiektów)

```python
from sklearn.neighbors import NearestNeighbors

# Define number of synthetic examples to generate
num_synthetic_examples = 18

# Define weight combinations for synthetic examples
weight_combinations = [(0.0, 1.0), (0.2, 0.8), (0.4, 0.6), (0.6, 0.4), (0.8, 0.2), (1.0, 0.0)]

# Generate synthetic examples for a few examples from each class
for class_idx in range(10):
    class_examples = np.where(train_labels.flatten() == class_idx)[0][99:102]  # Get indices of examples from the class

    for example_idx in class_examples:
        example_image = train_images[example_idx]
        example_label = train_labels[example_idx]

        # Reshape and normalize example image
        example_image_histogram = extract_color_histogram(np.array([example_image]))

        # Find nearest neighbor
        nn_model = NearestNeighbors(n_neighbors=9)
        nn_model.fit(train_images_histograms)
        _, nn_indices = nn_model.kneighbors(example_image_histogram)

        # Get the nearest neighbor image
        neighbor_image = train_images[nn_indices[0][2]]
        neighbor_label = train_labels[nn_indices[0][2]]

        plt.figure(figsize=(48, 24))
        plt.subplot(1, num_synthetic_examples + 2, 1)
        plt.imshow(example_image)
        plt.title('Example (Class {})'.format(class_idx))

        plt.subplot(1, num_synthetic_examples + 2, 2)
        plt.imshow(neighbor_image)
        plt.title('Nearest Neighbor (Class {})'.format(neighbor_label[0]))

        # Generate synthetic examples using different weight combinations
        for i, weights in enumerate(weight_combinations):
            synthetic_image = example_image * weights[0] + neighbor_image * weights[1]
            synthetic_label = example_label

            plt.subplot(1, num_synthetic_examples + 2, i + 3)
            plt.imshow(synthetic_image.astype(int))
            plt.title('Synthetic {} (Class {})'.format(i + 1, synthetic_label[0]))

        plt.tight_layout()
        plt.show()
```

## Część II: konwolucyjny ekstraktor cech

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.applications import ResNet50
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt

cnn_model = ResNet50(weights='imagenet', include_top=False, input_shape=(32, 32, 3))
cnn_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
cnn_model.fit(train_images, train_labels, epochs=20, batch_size=32)
```

```python
x_train = train_images
y_train = train_labels
x_test = test_images
y_test = test_labels

# 4. Usunięcie ostatnich warstw i przygotowanie ekstraktora cech
feature_extractor = tf.keras.models.Model(inputs=[cnn_model.input], outputs=[cnn_model.layers[-2].output])

# 5. Przepuszczenie danych przez ekstraktor cech
x_train_features = feature_extractor.predict(x_train).reshape(10000, 2048)
x_test_features = feature_extractor.predict(x_test).reshape(2000, 2048)
```

```python
# Klasyfikacja kNN dla różnych liczby sąsiadów i różnych podejść
k_values = [1, 3, 5, 7, 9]
approach_vector_accuracy = []
approach_histogram_accuracy = []

for k in k_values:
    print(k)
    vector_accuracy = kNN_classification_vector(x_train_features, y_train, x_test_features, y_test, k)
    approach_vector_accuracy.append(vector_accuracy)
```

```python
best_vector_accuracy = max(approach_vector_accuracy)
best_vector_k = k_values[approach_vector_accuracy.index(best_vector_accuracy)]

print("Najlepsze wyniki dla podejścia z ekstraktorem:")
print("Liczba sąsiadów:", best_vector_k)
print("Dokładność klasyfikacji:", best_vector_accuracy)
```

Uzyskane wyniki są nieco gorsze od poprzednich ale różnica nie jest duża - ok 5% gorzej niż wersja z wektorami poprzednio.

```python
# 6. Dopasowanie modelu kNN do przetworzonych danych
knn_model = NearestNeighbors(n_neighbors=best_vector_k)
knn_model.fit(x_train_features)

# 7. Odszukanie sąsiadów dla kilku przykładów
n_examples = 3
neighbors_indices = knn_model.kneighbors(x_test_features[:n_examples], return_distance=False)

# 8. Wyświetlenie przykładów i ich sąsiadów
fig, axes = plt.subplots(nrows=n_examples, ncols=2, figsize=(8, 4*n_examples))

for i in range(n_examples):
    # Wyświetlenie oryginalnego przykładu
    axes[i, 0].imshow(x_test[i])
    axes[i, 0].set_title(f'Example {i+1}')
    axes[i, 0].axis('off')

    # Wyświetlenie sąsiada
    neighbor_index = neighbors_indices[i][0]
    axes[i, 1].imshow(x_train[neighbor_index])
    axes[i, 1].set_title(f'Nearest Neighbor {neighbor_index}')
    axes[i, 1].axis('off')

plt.tight_layout()
plt.show()
```

Widać że tym razem klasyfikator nie skupia się tak bardzo na kolorystyce, a bardziej na podobnym ułożeniu obiektów.

```python
# 9. Przygotowanie SMOTE
smote_model = NearestNeighbors(n_neighbors=best_vector_k)
smote_model.fit(x_train_features)

weights = [(0.0, 1.0), (0.2, 0.8), (0.4, 0.6), (0.6, 0.4), (0.8, 0.2), (1.0, 0.0)]
n_samples = 3
n_classes = 10

fig, axes = plt.subplots(nrows=6, ncols=8, figsize=(30, 24))

for i in range(2):
    for j in range(3):
        # Wylosowanie próbki z danej klasy
        class_indices = np.where(y_train == i)
        random_index = np.random.choice(class_indices[0])
        example = x_train_features[random_index]

        # Znalezienie sąsiada dla próbki
        _, neighbor_index = smote_model.kneighbors(example.reshape(1, -1))
        neighbor = x_train[neighbor_index[0, 1]]

        # Wyświetlenie próbki i jej sąsiada
        axes[3*i + j, 0].imshow(x_train[random_index])
        axes[3*i + j, 0].set_title(f'Original {j+1}')
        axes[3*i + j, 0].axis('off')

        axes[3*i + j, 1].imshow(neighbor)
        axes[3*i + j, 1].set_title(f'Neighbor {neighbor_index[0, 0]}')
        axes[3*i + j, 1].axis('off')

        for idx, weight in enumerate(weights):
            # Generowanie próbki "pośredniej" na podstawie wagi
            interpolated_example = np.array(weight[0] * x_train[random_index] + weight[1] * neighbor).astype(int)

            # Wyświetlenie połączenia
            axes[3*i + j, 2+idx].imshow(interpolated_example)
            axes[3*i + j, 2+idx].set_title(f'Interpolated Example {j+1} ({weight[0]}, {weight[1]})')
            axes[3*i + j, 2+idx].axis('off')

plt.tight_layout()
plt.show()
```

Przez mniejszą dokłądność wyników widać większe rozbieżności w stosunku z poprzednią metodą, natomiast przez to, że bardziej brane jest pod uwagę umieszczenie obiektu na obrazku w niektórych przyładach (1 i 2) efekty połączenia są zadowalające.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors
from tensorflow.keras.models import Model,Sequential
from tensorflow.keras.layers import Input, Dense, Conv2D, MaxPooling2D, UpSampling2D, Flatten, Reshape, Conv2DTranspose
from keras.layers import Input, Dense,Conv2D,MaxPooling2D,UpSampling2D,BatchNormalization

# 1. Wytrenuj autoenkoder na zbiorze danych

x_train = train_images
y_train = train_labels
x_test = test_images
y_test = test_labels

# Przygotuj dane treningowe dla autoenkodera
x_train_autoencoder = x_train
y_train_autoencoder = y_train

# Normalizuj dane treningowe
# x_train_autoencoder = transform_images_to_vector(x_train_autoencoder.astype('float32') / 255.0)
x_train_autoencoder = x_train_autoencoder.astype('float32') / 255.0
x_test_autoencoder = x_test.astype('float32') / 255.0

# # Definiuj architekturę autoenkodera
input_shape = x_train_autoencoder.shape[1:]
latent_dim = 3
```

```python
input_img = Input(shape=input_shape)
encoded = Dense(latent_dim, activation='relu')(input_img)
decoded = Dense(input_shape[2], activation='sigmoid')(encoded)

autoencoder = Model(input_img, decoded)
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')

autoencoder.fit(x_train_autoencoder, x_train_autoencoder, epochs=10, batch_size=128)
```

```python
# 1.1. Przepuść zbiór przez autoenkoder i porównaj rekonstrukcje z oryginałami
x_train_reconstructed = autoencoder.predict(x_test_autoencoder)

plt.figure(figsize=(10, 4))
for i in range(5):
    plt.subplot(2, 5, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_test_autoencoder[i])
    plt.xlabel('Oryginał')

    plt.subplot(2, 5, i+6)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_train_reconstructed[i])
    plt.xlabel('Rekonstrukcja')
plt.show()
```
